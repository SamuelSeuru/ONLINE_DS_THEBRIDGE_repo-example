#### Existe alguna relacin estadsticamente significativa entre el sexo y la supervivencia en el Titanic?
 relacin entre dos variables categricas ->  vamos a emplear el test 

```python
tabla_contingencia = pd.crosstab(df_titanic['alive'], df_titanic['sex'])

tabla_contingencia
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabla_contingencia)


print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

### Interpretacin de resultados 

El test chi-2 tiene como hiptesus nula (o de partida) la independencia de las variables. El P-Value es menor de 0.05 : podemos pensar que existe una relacin estadsticamente significativa entre el sexo y la sobrevivencia.

### 4 Qu relacin hay entre el precio del pasaje y la supervivencia/fallecimiento?


```python
#observamos primero 
plot_grouped_histograms(df_titanic, cat_col="alive",num_col="fare", group_size= 2)
```

#### No tiene pinta de ser normal, ai que en ve de usar t-student test, usamos el test U de Man-Whitney

#### test U de Mann-Whitney -> una de las 2 variables es binaria



```python
#1 Separar los datos en dos grupos segn la variable categrica
grupo_a = df_titanic.loc[df_titanic.alive == "yes"]["fare"]
grupo_b = df_titanic.loc[df_titanic.alive == "no"]["fare"]
```

```python
#2 Aplicar la prueba y mostrar los resultados
u_stat, p_valor = mannwhitneyu(grupo_a, grupo_b)

print("Estadstico U:", u_stat)
print("Valor p:", p_valor)
```

##### Interpretacin del valor p (p-value)
p-value debajo de 0.05: rechazamos la hiptesis nula y entonces hay diferencia significativa (=hay relacin estadstica significativa entre la sobrevivencia y el precio)
