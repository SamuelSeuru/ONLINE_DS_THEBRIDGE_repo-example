#### ¿Existe alguna relación estadísticamente significativa entre el sexo y la supervivencia en el Titanic?
 relación entre dos variables categóricas ->  vamos a emplear el test χ²

```python
tabla_contingencia = pd.crosstab(df_titanic['alive'], df_titanic['sex'])

tabla_contingencia
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabla_contingencia)


print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

### Interpretación de resultados 

El test chi-2 tiene como hipótesus nula (o de partida) la independencia de las variables. El P-Value es menor de 0.05 : podemos pensar que existe una relación estadísticamente significativa entre el sexo y la sobrevivencia.

### 4 ¿Qué relación hay entre el precio del pasaje y la supervivencia/fallecimiento?


```python
#observamos primero 
plot_grouped_histograms(df_titanic, cat_col="alive",num_col="fare", group_size= 2)
```

#### No tiene pinta de ser normal, ai que en ve de usar t-student test, usamos el test U de Man-Whitney

#### test U de Mann-Whitney -> una de las 2 variables es binaria



```python
#1 Separar los datos en dos grupos según la variable categórica
grupo_a = df_titanic.loc[df_titanic.alive == "yes"]["fare"]
grupo_b = df_titanic.loc[df_titanic.alive == "no"]["fare"]
```

```python
#2 Aplicar la prueba y mostrar los resultados
u_stat, p_valor = mannwhitneyu(grupo_a, grupo_b)

print("Estadístico U:", u_stat)
print("Valor p:", p_valor)
```

##### Interpretación del valor p (p-value)
p-value debajo de 0.05: rechazamos la hipótesis nula y entonces hay diferencia significativa (=hay relación estadística significativa entre la sobrevivencia y el precio)

### #2.3

Crea una variable categorica "cat_vuelo" a partir de distancias, escoge los rangos de las categorías basándote en la distribución de valores de "distancia" (sugerencia entre 3 y 4 categorías). Extra: Reaiza ahora otra vez el análisis multivariante "cat_vuelo", "ingresos", "consumo_kg". ¿Ves algo diferente?

```python
distancia_min = df_air_jun["distancia"].min()
distancia_max = df_air_jun["distancia"].max()

limite1 = 2500
limite2 = 7700
limite3 = 11000

limites_categorias = [distancia_min, limite1, limite2, limite3, distancia_max]

nombres_categorias = ["Corta", "Media Corta", "Media Larga", "Larga"]


df_air_jun["cat_vuelo"] = pd.cut(df_air_jun["distancia"], bins=limites_categorias, labels=nombres_categorias, include_lowest=True)

```
